```python
class MyCircularQueue {

    private int[] queue;
    private int front;
    private int rear;
    private int size;
    
    public MyCircularQueue(int k) {
        this.queue = new int[k];
        this.front = 0;
        this.rear = -1;
        this.size = 0;
    }
    
    public boolean enQueue(int value) {
        if (isFull()) {
            return false;
        } else {
            this.rear += 1;
            this.rear %= queue.length;
            this.queue[this.rear] = value;
            this.size += 1;
            return true;
        }
    }
    
    public boolean deQueue() {
        if (isEmpty()) {
            return false;
        } else {
            this.front += 1;
            this.front %= queue.length;
            this.size -= 1;
            return true;
        }
    }
    
    public int Front() {
        if (isEmpty()) {
            return -1;
        } else {
            return this.queue[this.front];
        }
    }
    
    public int Rear() {
        if (isEmpty()) {
            return -1;
        } else {
            return this.queue[this.rear];
        }
    }
    
    public boolean isEmpty() {
        return this.size == 0;
    }
    
    public boolean isFull() {
        return this.size == queue.length;
    }
}
```

```java
public class ListNode {
      int val;
      ListNode next;
      ListNode() {}
      ListNode(int val) { this.val = val; }
      ListNode(int val, ListNode next) { this.val = val; this.next = next; }
}

class Solution {
    public ListNode removeElements(ListNode head, int val) {
        ListNode dummy = new ListNode(0);
        dummy.next = head;
        ListNode cur = dummy;
        
        while (cur.next != null) {
            if (cur.next.val == val)
                cur.next = cur.next.next;
            else
                cur = cur.next;
        }
        
        return dummy.next;
    }
}
```

```java
class Solution2 {
    public ListNode mergeTwoLists(ListNode list1, ListNode list2) {
        ListNode cur = new ListNode();
        ListNode dummy = cur;
        
        if (list1 == null) return list2;
        else if (list2 == null) return list1;
        
        ListNode p1 = list1;
        ListNode p2 = list2;
        
        while (p1 != null && p2 != null) {
            if (p1.val > p2.val) {
                cur.next = p2;
                cur = p2;
                p2 = p2.next;
            } else {
                cur.next = p1;
                cur = p1;
                p1 = p1.next;
            }
        }
        
        if (p1 == null)
            cur.next = p2;
        else
            cur.next = p1;
            
        
        return dummy.next;
    }
}
```

```java
class Solution3 {
    public ListNode middleNode(ListNode head) {
        ListNode slow = head;
        ListNode fast = head;
        
        while (fast.next != null) {
            slow = slow.next;
            if (fast.next.next == null)
                fast = fast.next;
            else
                fast = fast.next.next;
        }
        
        return slow;
    }
}
```

```java
public class Solution4 {
    public boolean hasCycle(ListNode head) {
        ListNode slow = head;
        ListNode fast = head;
        
        while (fast != null && fast.next != null) {
            slow = slow.next;
            if (fast.next == null)
                fast = fast.next;
            else
                fast = fast.next.next;
            
            if (slow == fast)
                return true;
        }
        
        return false;
    }
}
```

```java
public class Solution5 {
    public ListNode getIntersectionNode(ListNode headA, ListNode headB) {
        ListNode p1 = headA;
        ListNode p2 = headB;
        
        while (p1 != p2) {
            p1 = (p1 == null ? headB : p1.next);
            p2 = (p2 == null ? headA : p2.next);
        }
        
        return p1;
    }
}
```

```java
class Solution6 {
    public ListNode reverseList(ListNode head) {
        
        if (head == null) return null;
        if (head.next == null) return head;
        
        ListNode prev = head;
        ListNode cur = prev.next;
        ListNode next = cur.next;
        
        while (cur != null) {
            if (prev == head)
                prev.next = null;
            cur.next = prev;
            prev = cur;
            cur = next;
            if (next == null)
                next = null;
            else
                next = cur.next;
        }
        return prev;
    }
}
```

```java
class Solution7 {
    public ListNode addTwoNumbers(ListNode p1, ListNode p2) {
        ListNode dummy = new ListNode();
        ListNode cur = dummy;
        int carry = 0;
        
        while (p1 != null || p2 != null) {
            int p1Val = p1 == null ? 0 : p1.val;
            int p2Val = p2 == null ? 0 : p2.val;
            
            int remain = (p1Val+p2Val+carry) % 10;
            carry = (p1Val+p2Val+carry) / 10;
            
            ListNode newList = new ListNode(remain);
            cur.next = newList;
            cur = newList;
            
            p1 = p1 == null ? null : p1.next;
            p2 = p2 == null ? null : p2.next;
        }
        
        if (carry != 0) {
            ListNode newList = new ListNode(carry);
            cur.next = newList;
            cur = newList;
        }
        
        return dummy.next;
    }
}
```

```java
class Solution9 {
    public ListNode removeNthFromEnd(ListNode head, int n) {
        ListNode dummy = new ListNode();
        dummy.next = head;
        ListNode slow = dummy;
        ListNode fast = dummy;
        
        for (int i = 0; i < n; i++)
            fast = fast.next;
        
        while (fast.next != null) {
            slow = slow.next;
            fast = fast.next;
        }
        
        slow.next = slow.next.next;
        return dummy.next;
    }
}
```

```java
class Solution10 {
    public ListNode partition(ListNode head, int x) {
        ListNode LessNode = new ListNode();
        ListNode GreaterNode = new ListNode();
        
        ListNode LessNodeHead = LessNode;
        ListNode GreaterNodeHead = GreaterNode;
        
        ListNode cur = head;
        
        while (cur != null) {
            if (cur.val < x) {
                LessNode.next = cur;
                LessNode = cur;
                cur = cur.next;
            } else {
                GreaterNode.next = cur;
                GreaterNode = cur;
                cur = cur.next;
            }
        }
        GreaterNode.next = null;
        LessNode.next = GreaterNodeHead.next;
        return LessNodeHead.next;
    }
}
```

```java
class Solution11 {
    public void reorderList(ListNode head) {
        
        if (head == null || head.next == null) return;
        
        ListNode dummy = new ListNode();
        ListNode cur = dummy;
        
        
        boolean flag = false;
        
        ListNode middle = MiddleNode(head);
        ListNode reversedHead = reverse(middle.next);
        middle.next = null;
        
        while (cur != middle) {
            if (flag) {
                cur.next = reversedHead;
                cur = reversedHead;
                reversedHead = reversedHead.next;
                flag = false;
            } else {
                cur.next = head;
                cur = head;
                head = head.next;
                flag = true;
            }
        }
        
        if (reversedHead != null)
            cur.next= reversedHead;
        
    }
    
    public ListNode MiddleNode(ListNode head) {
        ListNode slow = head, fast = head.next;
        while (fast != null && fast.next != null) {
            slow = slow.next;
            fast = fast.next.next;
        }
        
        return slow;
    }
    
    public ListNode reverse(ListNode head) {
        
        if (head == null) return null;
        if (head.next == null) return head;
        
        ListNode prev = head;
        ListNode cur = prev.next;
        ListNode next = cur.next;
        
        while (cur != null) {
            if (prev == head)
                prev.next = null;
            cur.next = prev;
            prev = cur;
            cur = next;
            if (next == null)
                next = null;
            else
                next = cur.next;
        }
        return prev;
    }
}
```

```java
class Node {
    int val;
    Node next;
    Node random;

    public Node(int val) {
        this.val = val;
        this.next = null;
        this.random = null;
    }
}

class Solution12 {
    public Node copyRandomList(Node head) {
        Map<Node, Node> map = new HashMap<>();
        Node cur1 = head;

        Node dummy = new Node(0);
        Node cur2 = dummy;

        while (cur1 != null) {
            Node newNode = new Node(cur1.val);
            map.put(cur1, newNode);

            cur2.next = newNode;
            cur2 = cur2.next;
            cur1 = cur1.next;
        }

        cur2 = dummy.next;
        cur1 = head;
        
        while (cur1 != null) {
            cur2.random = map.get(cur1.random);
            cur1 = cur1.next;
            cur2 = cur2.next;
        }
        
        return dummy.next;
    }
}
```

```java
class LRUCache extends LinkedHashMap<Integer, Integer>{

    private final int capacity;

    public LRUCache(int capacity) {
        super(capacity, 0.75f, true);
        this.capacity = capacity;
    }

    public int get(int key) {
        return super.getOrDefault(key, -1);
    }

    @Override
    protected boolean removeEldestEntry(Map.Entry eldest) {
        return size() > capacity;
    }

}
```
